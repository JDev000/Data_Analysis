```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
location = 'Data - Sheet.csv'
```

```python
df = pd.read_csv(location)
```

```python
print(df)
```

```python
major_counts = df.value_counts()

plt.pie(major_counts, labels=major_counts.index, autopct='%1.1f%%')
plt.title('Basic Distribution')
plt.show()
```
